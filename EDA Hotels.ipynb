```python
cancelled_perc=df['is_canceled'].value_counts(normalize=True)
print(cancelled_perc)
```

```python
plt.figure(figsize=(5,4))
plt.title("Reservation status")
plt.bar(['Not Canceled' , 'Canceled'] ,df['is_canceled'].value_counts())
```

```python
# Above we can see that the cancelation percent is more than half
```

```python
plt.figure(figsize= (6,4))
ax1=sns.countplot(x= 'hotel' , hue = 'is_canceled' , data=df , palette='Blues')
plt.legend(['not_cancelled' , 'canceled'])
```

```python
#Above we can see city hotels have much higher cancellation rates
```

```python
resort_hotel=df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
#Above we can see for Resort hotels 27.9% are being canceled
```

```python
city_hotel=df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
#Above we can see 41.7% are getting cancelled in City Hotels
```

```python
resort_hotel=resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel=city_hotel.groupby('reservation_status_date')[['adr']].mean()

plt.figure(figsize=(20,10))
plt.title('Avg Daily rate in city and resort hotel', fontsize=30)
plt.plot(resort_hotel.index,resort_hotel['adr'],label='Resort Hotel' )
plt.plot(city_hotel.index,city_hotel['adr'],label='City Hotel' )
plt.legend(fontsize=20)
plt.show()
```

```python
df['month']=df['reservation_status_date'].dt.month
plt.figure(figsize=(14,7))
ax1=sns.countplot(x='month' , hue='is_canceled',data=df)
plt.legend(['not_canceled' ,'canceled'])
```

```python
plt.figure(figsize=(15, 8))
plt.title('ADR per month', fontsize=30)
sns.barplot(x='month', y='adr', data=df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index())
plt.show()
```

```python
#Hypothesis : January has the most cancelled reservation status as the prices are high in Jan according to the above graphs.
```

```python
cancelled_data=df[df['is_canceled']==1]
top_10_countries=cancelled_data['country'].value_counts()[:10]
plt.figure(figsize=(8,8))
plt.pie(top_10_countries , autopct='%.2f',labels=top_10_countries.index)
plt.show()
```

```python
#Portugal has highest cancellation rates
```
